```python
from googleapiclient.discovery import build
import pandas as pd

# Thay th·∫ø 'YOUR_API_KEY' b·∫±ng API key b·∫°n ƒë√£ t·∫°o
API_KEY = ....

# T·∫°o m·ªôt ƒë·ªëi t∆∞·ª£ng API client cho YouTube Data API v3
youtube = build('youtube', 'v3', developerKey=API_KEY)

# H√†m ƒë·ªÉ g·ªçi API v√† l·∫•y d·ªØ li·ªáu video
def get_videos(youtube, max_results=200, region_code='VN'):
    video_data = []
    next_page_token = None

    while len(video_data) < max_results:
        request = youtube.videos().list(
            part='snippet,contentDetails,statistics',
            chart='mostPopular',
            regionCode=region_code,
            maxResults=50,
            pageToken=next_page_token
        )

        response = request.execute()

        for item in response['items']:
            video_info = {
                'Video_id': item['id'],
                'Published_at': item['snippet']['publishedAt'],
                'Video_title': item['snippet']['title'],
                'Channel_name': item['snippet']['channelTitle'],
                'Channel_id': item['snippet']['channelId'],
                'Category_id': item['snippet']['categoryId'],
                'Live_Broadcast_Content': item['snippet']['liveBroadcastContent'],
                'Views': item['statistics'].get('viewCount', 'N/A'),
                'Likes': item['statistics'].get('likeCount', 'N/A'),
                'Dislikes': item['statistics'].get('dislikeCount', 'N/A'),
                'Favorite': item['statistics'].get('favoriteCount', 'N/A'),
                'Comment': item['statistics'].get('commentCount', 'N/A')
            }
            video_data.append(video_info)

        next_page_token = response.get('nextPageToken')

        if not next_page_token:
            break

    return video_data

# G·ªçi h√†m ƒë·ªÉ l·∫•y d·ªØ li·ªáu
video_data = get_videos(youtube, max_results=1000, region_code='VN')

# T·∫°o dataframe t·ª´ danh s√°ch video_data
df = pd.DataFrame(video_data)

df.head()
```

```python
df['Published_date'] = df['Published_at'].apply(lambda x: x.split('T')[0])
df['Publised_time'] = df['Published_at'].apply(lambda x: x.split('T')[1].replace('Z', ''))
df.drop(columns=['Published_at'], inplace=True)
df.head()
```

```python
df.info()
```

```python
# Change type data
df['Category_id'] = df['Category_id'].astype(int)
df['Views'] = df['Views'].astype(int)
df['Likes'] = pd.to_numeric(df['Likes'], errors='coerce').astype('Int32')
df['Dislikes'] = pd.to_numeric(df['Dislikes'], errors='coerce').astype('Int32')
df['Favorite'] = pd.to_numeric(df['Favorite'], errors='coerce').astype('Int32')
df['Comment'] = pd.to_numeric(df['Comment'], errors='coerce').astype('Int32')
df['Published_date'] = pd.to_datetime(df['Published_date'], errors='coerce')
df['Publised_time'] = pd.to_datetime(df['Publised_time'], format='%H:%M:%S').dt.time
df.head()
```
